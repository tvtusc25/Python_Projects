#### Check overfitting with $p = 7$ polynomial regression model

```python
# Fit a polynomial regression model with p=7 on the fit set
poly = 7
fit_lr = linear_regression.LinearRegression(fit_data)
fit_lr.poly_regression(['X'], 'Y', poly)

# Create a new LinearRegression object with the validation data
val_lr = linear_regression.LinearRegression(val_data)
val_lr.poly_regression(['X'], 'Y', poly)

# Copy over the fitted coefficients to the validation data
val_lr.initialize('X', 'Y', fit_lr.get_fitted_slope(), fit_lr.get_fitted_intercept(), poly)

# Create scatterplots
fit_lr.scatter('X', 'Y', 'Fit Set')
val_lr.scatter('X', 'Y', 'Validation Set')

# Compute and print MSE for both sets
fit_mse = fit_lr.compute_mse()
val_mse = val_lr.compute_mse()
print('Fit Set MSE:', fit_mse)
print('Validation Set MSE:', val_mse)
```

** 2: I created a method for calculating linear regression using normal equations, and I implemented matrix condition number. Calculating linear regression using normal equations had the same result for R^2 and MSE as using scipy, which can be seen by comparing the plot below to the one in 3c. Hence, it fits equally well. I also calculated the matrix number for the data set poly (x vs. y) and iris (petal width vs. sepal width).
```python
normal_filename = 'data/poly_data.csv'
normal_data = data.Data(normal_filename)
lin_reg = linear_regression.LinearRegression(normal_data)
lin_reg.linear_regression_normal(['X'], 'Y')
print("MSE: ", lin_reg.compute_mse())
print("Condition Number: ",lin_reg.get_cond_x())
lin_reg.scatter('X', 'Y', 'Linear Regression Using Normal Equations on Poly!')
lin_reg.show()

iris_filename = 'data/iris.csv'
iris_data = data.Data(iris_filename)
lin_reg = linear_regression.LinearRegression(iris_data)
lin_reg.linear_regression_normal(['sepal_length'], 'petal_width')
print("MSE: ", lin_reg.compute_mse())
print("Condition Number: ",lin_reg.get_cond_x())
lin_reg.scatter('sepal_length', 'petal_width', 'Linear Regression Using Normal Equations on Iris!')
lin_reg.show()
```
